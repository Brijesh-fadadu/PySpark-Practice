```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('lr').getOrCreate()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
training = spark.read.format('libsvm').load('/content/drive/MyDrive/Spark with Python/Python-and-Spark-for-Big-Data-master/Spark_for_Machine_Learning/Linear_Regression/sample_linear_regression_data.txt')
```

```python
training.show()
```

```python
lr = LinearRegression(featuresCol='features', labelCol='label', predictionCol='prediction')
```

```python
lrModel = lr.fit(training)
```

```python
lrModel.coefficients
```

```python
lrModel.intercept
```

```python
training_summary = lrModel.summary
```

```python
training_summary.rootMeanSquaredError
```

```python
all_data = spark.read.format('libsvm').load('/content/drive/MyDrive/Spark with Python/Python-and-Spark-for-Big-Data-master/Spark_for_Machine_Learning/Linear_Regression/sample_linear_regression_data.txt')
```

```python
train_data, test_data = all_data.randomSplit([0.7,0.3])
```

```python
train_data.describe().show()
```

```python
test_data.describe().show()
```

```python
correct_model = lr.fit(train_data)
```

```python
test_results = correct_model.evaluate(test_data)
```

```python
test_results.rootMeanSquaredError
```

```python
unlabeled_data = test_data.select('features')
```

```python
unlabeled_data.show()
```

```python
prediction = correct_model.transform(unlabeled_data)
```

```python
prediction.show()
```
